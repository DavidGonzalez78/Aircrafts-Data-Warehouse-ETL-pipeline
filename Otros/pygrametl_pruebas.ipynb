```python
print("\n" + "="*50)
print("EXPLORANDO DATOS DE EJEMPLO")
print("="*50)

# 1. Ver estructura y datos de AIMS.FLIGHTS
print("\n AIMS.FLIGHTS (primeras 3 filas):")
cursor.execute("SELECT * FROM \"AIMS\".\"flights\" LIMIT 5")
flights_data = cursor.fetchall()
for fila in flights_data:
    print(fila)

# Ver columnas de flights
cursor.execute("""
    SELECT column_name, data_type 
    FROM information_schema.columns 
    WHERE table_schema = 'AIMS' AND table_name = 'flights'
    ORDER BY ordinal_position
""")
print("\nColumnas de AIMS.flights:")
for col, tipo in cursor.fetchall():
    print(f"- {col} ({tipo})")

```

```python
# 2. Ver estructura y datos de AMOS.maintenanceevents  
print("\n\nAMOS.MaintenanceEvents (primeras 3 filas):")
cursor.execute("SELECT * FROM \"AMOS\".\"maintenanceevents\" LIMIT 5")
maint_data = cursor.fetchall()
for fila in maint_data:
    print(fila)
```

```python
# 3. Buscar tablas de LOGBOOK (para los reportes)
print("\n\nBUSCANDO TABLAS DE LOGBOOK/REPORTES:")
cursor.execute("""
    SELECT table_name 
    FROM information_schema.tables 
    WHERE table_schema = 'AMOS' 
    AND table_name ILIKE '%log%' OR table_name ILIKE '%report%'
""")
log_tables = cursor.fetchall()
print("Posibles tablas de logbook:")
for tabla in log_tables:
    print(f"- {tabla[0]}")
```
